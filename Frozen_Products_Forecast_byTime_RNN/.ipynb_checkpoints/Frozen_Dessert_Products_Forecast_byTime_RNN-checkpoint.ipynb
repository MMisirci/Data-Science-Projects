```python
## Some Useful Functions

###############################################################################

def df_general_view(df):
    print(colored("Shape:", attrs=['bold']), df.shape,'\n',
          f"There is ", df.shape[0], " observation and ", df.shape[1], " columns in the dataset.", '\n',
          colored('-'*79, 'red', attrs=['bold']),
          colored("\nInfo:\n", attrs=['bold']), sep='')
    print(df.info(), '\n', 
          colored('-'*79, 'red', attrs=['bold']), sep='')
    print(colored("Number of Uniques:\n", attrs=['bold']), df.nunique(),'\n',
          colored('-'*79, 'red', attrs=['bold']), sep='')
    if missing_values(df).empty:
        print(colored("Missing Values: ", attrs=['bold']), "There is no missing value in the given dataframe",'\n',
              colored('-'*79, 'red', attrs=['bold']), sep='')
    else:
        print(colored("Missing Values:\n", attrs=['bold']), missing_values(df),'\n',
              colored('-'*79, 'red', attrs=['bold']), sep='')

###############################################################################

def missing_values(df):
    missing_number = df.isnull().sum().sort_values(ascending=False)
    missing_percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
    missing_values = pd.concat([missing_number, missing_percent], axis=1, keys=['Missing_Number', 'Missing_Percent'])
    missing_values=missing_values[missing_values['Missing_Number']>0]
    return missing_values

###############################################################################

def duplicate_values(df):
    print(colored("Duplicate check...", attrs=['bold']), sep='')
    duplicate_values = df.duplicated(subset=None, keep='first').sum()
    if duplicate_values > 0:
        df.drop_duplicates(keep='first', inplace=True)
        print(duplicate_values, colored("Duplicates were dropped!"),'\n',
              colored('-'*79, 'red', attrs=['bold']), sep='')
    else:
        print(colored("There are no duplicates"),'\n',
              colored('-'*79, 'red', attrs=['bold']), sep='') 
    
###############################################################################
```

# Preprocessing of Data

## Train Test Split

```python
len(df)
```

```python
test_size = 72

test_ind = len(df) - test_size
```

```python
train = df.iloc[:test_ind]
test = df.iloc[test_ind:]
```

```python
len(test)
```

## Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
train_scaled = scaler.fit_transform(train)
```

```python
test_scaled = scaler.transform(test)
```

## Time Series Generator

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
# help(TimeseriesGenerator)
```

```python
length = 24
batch_size = 1
generator = TimeseriesGenerator(train_scaled, train_scaled, length = length, batch_size = batch_size)
```

# Modelling & Model Performance

## with "tanh" activation function

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from keras.layers import Dropout
```

```python
n_features = 1
```

```python
model = Sequential()
model.add(LSTM(42, activation = 'tanh', input_shape = (length, n_features)))
#model.add(Dropout(0.2))
model.add(Dense(1))
model.compile(optimizer = 'adam', loss = 'mse')
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
```

```python
early_stop = EarlyStopping(monitor = 'val_loss', patience = 2, mode='min', verbose=1)
#early_stop= EarlyStopping(monitor='val_loss', min_delta=1e-4, patience=10,restore_best_weights=True)
learning_rate_reduction = ReduceLROnPlateau(monitor='val_loss', patience=7, verbose=10, factor=0.5, min_lr=0.0000001)
early_stopping = EarlyStopping(monitor='val_loss', min_delta=1e-10, patience=25,restore_best_weights=True)
callbacks_list2=[learning_rate_reduction,early_stopping]
```

```python
validation_generator = TimeseriesGenerator(test_scaled, test_scaled, length = length, batch_size = 1)
```

```python
len(validation_generator)
```

```python
model.fit_generator(generator, epochs = 100,
                    validation_data = validation_generator,
                    callbacks = callbacks_list2)
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot()
```

### Evaluation on Test Data

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python

def eval_metrics(actual, pred):
    rmse = np.sqrt(mean_squared_error(actual, pred))
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    score = r2_score(actual, pred)
    return print("r2_score:", score, "\n","mae:", mae, "\n","mse:",mse, "\n","rmse:",rmse)
```

```python
test_predictions = []

first_eval_batch = train_scaled[-length:]

current_batch = first_eval_batch.reshape((1, length, n_features))

for i in range(len(test)):
    
    # get prediction 1 time stamp ahead
    current_pred = model.predict(current_batch)[0]
    
    # store prediction
    test_predictions.append(current_pred) 
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:, 1:, :], [[current_pred]], axis=1)
```

# Retrain and Forecasting

Select activation function providing a better score, create your final model with full data, forecast for the next 12 months, and plot this forecast.

```python
full_scaler = StandardScaler()
scaled_full_data = full_scaler.fit_transform(df)
```

```python
length = 24 
generator = TimeseriesGenerator(scaled_full_data, scaled_full_data, length = length, batch_size = 1)
```

```python
model = Sequential()

model.add(LSTM(78, activation = 'relu', input_shape = (length, n_features)))
#model.add(Dropout(0.2)) 
model.add(Dense(1)) 

model.compile(optimizer = 'adam', loss = 'mse')
model.summary()


model.fit_generator(generator, epochs=64)
```

```python
model.save('modelfinal.h5')
```

```python
scaled_full_data.shape
```

```python
scaled_full_data[-length:].shape
```

```python
forecast = []
# Replace periods with whatever forecast length you want
periods = 24

first_eval_batch = scaled_full_data[-length:] 
current_batch = first_eval_batch.reshape((1, length, n_features))

for i in range(periods):
    
    # get prediction 1 time stamp ahead ([0] is for grabbing just the number instead of [array])
    current_pred = model.predict(current_batch)[0]
    
    # store prediction
    forecast.append(current_pred) 
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:, 1:, :], [[current_pred]], axis = 1)
```

```python
forecast = scaler.inverse_transform(forecast)
```
